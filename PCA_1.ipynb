# Dimensionality Reduction on the Iris Dataset

## Introduction
This Colab notebook will explore several dimensionality reduction techniques applied to the classic Iris dataset. These techniques are essential in machine learning for feature extraction, visualization, and to enhance the performance of models.

## Dataset
The Iris dataset consists of 150 samples of iris flowers, each with four features: sepal length, sepal width, petal length, and petal width. It's a multi-class dataset with three different iris species.

## Techniques Covered
PCA (Principal Component Analysis)
SVD (Singular Value Decomposition)
MDS (Multidimensional Scaling)
ISOMap (Isometric Mapping)
LLE (Locally Linear Embedding)
UMAP (Uniform Manifold Approximation and Projection)
t-SNE (t-Distributed Stochastic Neighbor Embedding)


```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

###1. PCA (Principal Component Analysis)
PCA is a statistical procedure that uses an orthogonal transformation to convert a set of observations of possibly correlated variables into a set of linearly uncorrelated variables called principal components. It's widely used for exploratory data analysis and for making predictive models. PCA is particularly powerful in dealing with multicollinearity and in compressing the dataset while retaining most of the information.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Applying PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualizing PCA
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.show()
```

### 2. SVD (Singular Value Decomposition)
Singular Value Decomposition is a matrix factorization technique that decomposes a matrix into three other matrices. In the context of dimensionality reduction, SVD is used to reduce the number of features while retaining the essential information. It's particularly useful in applications like image compression and noise reduction.

```python
from sklearn.decomposition import TruncatedSVD

# Applying SVD
svd = TruncatedSVD(n_components=2)
X_svd = svd.fit_transform(X_scaled)

# Visualizing SVD
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y)
plt.xlabel('SVD Component 1')
plt.ylabel('SVD Component 2')
plt.title('SVD of Iris Dataset')
plt.show()
```

### 3. MDS (Multidimensional Scaling)
Multidimensional Scaling is a distance-preserving manifold learning method. It aims to place each object in N-dimensional space such that the between-object distances are preserved as well as possible. MDS is particularly useful for visualizing the level of similarity of individual cases in a dataset.

```python
from sklearn.manifold import MDS

# Applying MDS
mds = MDS(n_components=2)
X_mds = mds.fit_transform(X_scaled)

# Visualizing MDS
plt.scatter(X_mds[:, 0], X_mds[:, 1], c=y)
plt.xlabel('MDS Component 1')
plt.ylabel('MDS Component 2')
plt.title('MDS of Iris Dataset')
plt.show()
```

###4. ISOMap (Isometric Mapping)
Isometric Mapping, or ISOMap, is a nonlinear dimensionality reduction method that is particularly good at unraveling a data manifold. The technique aims to preserve the geodesic distances in the lower-dimensional space, making it an excellent tool for unfolding twisted and folded data structures.

```python
from sklearn.manifold import Isomap

# Applying ISOMap
isomap = Isomap(n_components=2)
X_isomap = isomap.fit_transform(X_scaled)

# Visualizing ISOMap
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=y)
plt.xlabel('ISOMap Component 1')
plt.ylabel('ISOMap Component 2')
plt.title('ISOMap of Iris Dataset')
plt.show()
```

### 5. LLE (Locally Linear Embedding)
Locally Linear Embedding is another nonlinear technique used for mapping high-dimensional data to a lower-dimensional space. It works by preserving local relationships within the data, making it ideal for exploring the local structure and features of data.

```python
from sklearn.manifold import LocallyLinearEmbedding

# Applying LLE
lle = LocallyLinearEmbedding(n_components=2)
X_lle = lle.fit_transform(X_scaled)

# Visualizing LLE
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=y)
plt.xlabel('LLE Component 1')
plt.ylabel('LLE Component 2')
plt.title('LLE of Iris Dataset')
plt.show()
```

### 6. UMAP (Uniform Manifold Approximation and Projection)
UMAP is a relatively new, non-linear dimensionality reduction technique that is particularly effective for visualizing clusters or groups in high-dimensional data. It's similar to t-SNE but often provides better runtime performance and can scale more easily to larger datasets.

```python
import umap

# Applying UMAP
umap_model = umap.UMAP(n_components=2)
X_umap = umap_model.fit_transform(X_scaled)

# Visualizing UMAP
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y)
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.title('UMAP of Iris Dataset')
plt.show()
```

### 7. t-SNE (t-Distributed Stochastic Neighbor Embedding)
t-SNE is a highly popular tool for high-dimensional data visualization. It works by converting similarities between data points to joint probabilities and minimizing the Kullback-Leibler divergence between the joint probabilities of the low-dimensional embedding and the high-dimensional data. t-SNE is particularly effective for creating compelling two- or three-dimensional maps of high-dimensional data.

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Applying t-SNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X_scaled)

# Visualizing t-SNE
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE of Iris Dataset')
plt.show()
```
